### Discrete Data Preparation – Weight of Evidence (WoE) and Information Value (IV)

**Weight of Evidence (WoE)** and **Information Value (IV)** are powerful techniques used in data preparation for predictive modeling, especially in credit risk modeling. WoE helps in understanding the predictive power of an independent variable in relation to a dependent variable, while IV ranks variables based on their predictive significance.

### Weight of Evidence (WoE)

**Definition**: Weight of Evidence quantifies the extent to which an independent variable can predict a dependent variable, typically in binary classification tasks such as default vs. non-default in credit risk. WoE is particularly useful when transforming continuous variables into categorical bins for improved interpretability in logistic regression models.

**Key Interpretation**:
- **Positive WoE**: Indicates that the distribution of “Goods” (e.g., non-defaults) is greater than the distribution of “Bads” (e.g., defaults).
- **Negative WoE**: Indicates that the distribution of “Goods” is less than the distribution of “Bads.”
- **Calculation Rule of Thumb**: The logarithmic function in WoE means:
  - If the log is greater than 1, the WoE is positive.
  - If the log is less than 1, the WoE is negative.

#### Formula for Weight of Evidence (WoE)
The WoE for each bin \( i \) is calculated as:

WoE= ln {{Distribution of Goods}/{{Distribution of Bads}}

where:
- **Distribution of Goods** is the proportion of non-events in bin \( i \).
- **Distribution of Bads** is the proportion of events in bin \( i \).

### Classification Techniques in WoE Calculation
WoE can be computed using two main binning techniques:
1. **Fine Classing**: Creates 10 to 20 bins/groups for a continuous variable, allowing for a detailed calculation of WoE and IV. Each bin represents a small range of the continuous variable, capturing subtle changes in predictive power.
2. **Coarse Classing**: Groups adjacent bins with similar WoE scores, reducing the number of bins. This technique provides a simplified and aggregated view of the variable, reducing noise and making the model more interpretable.

### Information Value (IV)

**Definition**: Information Value measures how predictive an independent variable is in explaining the dependent variable, often ranking variables in terms of their importance in the model. The higher the IV, the stronger the variable’s ability to separate “Goods” from “Bads.”

#### Formula for Information Value (IV)
The Information Value for a variable is calculated by summing the contributions of each bin \( i \):

IV = ∑ (% of non-events - % of events) * WOE

where:
- **% of Non-Events** for bin \( i \): The percentage of non-events (Goods) in that bin.
- **% of Events** for bin \( i \): The percentage of events (Bads) in that bin.
- **WoE** for bin \( i \): The weight of evidence for that bin.

### IV Categories and Predictive Power

The IV score helps in identifying the strength of a variable in predicting the outcome, with the following general thresholds:

- **IV < 0.02**: Not useful for prediction.
- **0.02 ≤ IV < 0.1**: Weak predictive power.
- **0.1 ≤ IV < 0.3**: Medium predictive power.
- **0.3 ≤ IV < 0.5**: Strong predictive power.
- **IV > 0.5**: Suspiciously high predictive power, suggesting potential overfitting or data leakage.

Using WoE and IV, we can identify the most relevant predictors, transform continuous variables effectively, and simplify the model-building process for interpretability and compliance with regulatory requirements. Both metrics are essential in credit risk modeling, especially when evaluating the likelihood of default, as they provide transparency and consistency in scoring model variables.

```python
# WoE function for discrete unordered variables
def woe_discrete(loan_data, discrete_variable_name, good_bad_variable_loan_data):
    # Step 1: Combine the loan_data containing the discrete variable and the target variable (good_bad)
    loan_data = pd.concat([loan_data[discrete_variable_name], good_bad_variable_loan_data], axis=1)

    # Step 2: Create two groupings:
    # - Count of observations for each category of the discrete variable.
    # - Mean of the target variable (proportion of "good" loans) for each category.
    loan_data = pd.concat([loan_data.groupby(loan_data.columns.values[0], as_index=False)[loan_data.columns.values[1]].count(),
                           loan_data.groupby(loan_data.columns.values[0], as_index=False)[loan_data.columns.values[1]].mean()], axis=1)

    # Step 3: Keep only the necessary columns from the resulting loan_data: category, count of observations, and proportion of good loans.
    loan_data = loan_data.iloc[:, [0, 1, 3]]

    # Step 4: Rename the columns for clarity:
    # - First column: the discrete variable.
    # - 'n_obs': number of observations (loans) in each category.
    # - 'prop_good': proportion of good loans (1's) in each category.
    loan_data.columns = [loan_data.columns.values[0], 'n_obs', 'prop_good']

    # Step 5: Calculate the proportion of observations for each category.
    loan_data['prop_n_obs'] = loan_data['n_obs'] / loan_data['n_obs'].sum()

    # Step 6: Calculate the number of good loans (n_good) for each category.
    loan_data['n_good'] = loan_data['prop_good'] * loan_data['n_obs']

    # Step 7: Calculate the number of bad loans (n_bad) for each category.
    loan_data['n_bad'] = (1 - loan_data['prop_good']) * loan_data['n_obs']

    # Step 8: Calculate the proportion of total good loans for each category.
    loan_data['prop_n_good'] = loan_data['n_good'] / loan_data['n_good'].sum()

    # Step 9: Calculate the proportion of total bad loans for each category.
    loan_data['prop_n_bad'] = loan_data['n_bad'] / loan_data['n_bad'].sum()

    # Step 10: Calculate the WoE (Weight of Evidence) for each category using the formula: ln(prop_n_good / prop_n_bad).
    loan_data['WoE'] = np.log(loan_data['prop_n_good'] / loan_data['prop_n_bad'])

    # Step 11: Sort the categories by WoE to see the order of risk from lowest to highest.
    loan_data = loan_data.sort_values(['WoE'])

    # Step 12: Reset the index for better readability after sorting.
    loan_data = loan_data.reset_index(drop=True)

    # Step 13: Calculate the absolute difference in the proportion of good loans between consecutive categories.
    loan_data['diff_prop_good'] = loan_data['prop_good'].diff().abs()

    # Step 14: Calculate the absolute difference in WoE between consecutive categories.
    loan_data['diff_WoE'] = loan_data['WoE'].diff().abs()

    # Step 15: Calculate the Information Value (IV) for the variable.
    # IV measures the predictive power of the variable.
    loan_data['IV'] = (loan_data['prop_n_good'] - loan_data['prop_n_bad']) * loan_data['WoE']

    # Step 16: Sum the IV values to get the total IV for the entire variable.
    loan_data['IV'] = loan_data['IV'].sum()

    # Step 17: Return the final dataframe with all the computed values.
    return loan_data

```

```python
# Below we define a function that takes 2 arguments: a dataframe and a number.
# The number parameter has a default value of 0.
# This means that if we call the function and omit the number parameter, it will be executed with it having a value of 0.
# The function displays a graph.
def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
    x = np.array(df_WoE.iloc[:, 0].apply(str))
    # Turns the values of the column with index 0 to strings, makes an array from these strings, and passes it to variable x.
    y = df_WoE['WoE']
    # Selects a column with label 'WoE' and passes it to variable y.
    plt.figure(figsize=(18, 6))
    # Sets the graph size to width 18 x height 6.
    plt.plot(x, y, marker = 'o', linestyle = '--', color = 'k')
    # Plots the datapoints with coordiantes variable x on the x-axis and variable y on the y-axis.
    # Sets the marker for each datapoint to a circle, the style line between the points to dashed, and the color to black.
    plt.xlabel(df_WoE.columns[0])
    # Names the x-axis with the name of the column with index 0.
    plt.ylabel('Weight of Evidence')
    # Names the y-axis 'Weight of Evidence'.
    plt.title(str('Weight of Evidence by ' + df_WoE.columns[0]))
    # Names the grapth 'Weight of Evidence by ' the name of the column with index 0.
    plt.xticks(rotation = rotation_of_x_axis_labels)
    # Rotates the labels of the x-axis a predefined number of degrees.
```

# WOE/IV for ordered discrete and continuous variables

```python
# WoE function for ordered discrete and continuous variables
def woe_ordered_continuous(loan_data, discrete_variable_name, good_bad_variable_df):
    # Combine the selected discrete variable and the good/bad target variable into a new DataFrame
    df = pd.concat([loan_data[discrete_variable_name], good_bad_variable_df], axis=1)
    
    # Group by the discrete variable and calculate counts and mean of the target variable
    df = pd.concat([df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1)
    
    # Select the relevant columns
    df = df.iloc[:, [0, 1, 3]]
    
    # Rename columns for clarity
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    
    # Calculate proportions of observations and good/bad counts
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    
    # Calculate Weight of Evidence (WoE)
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    
    # Calculate differences in proportions and WoE
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    
    # Calculate Information Value (IV)
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    
    # Return the resulting DataFrame with calculated values
    return df
```
